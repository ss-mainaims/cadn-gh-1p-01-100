```python
import argparse
from pytube import YouTube
from pytube import Playlist
import threading
import time
```

```python
def time_logger(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Function download_batch'{func.__name__}' took {execution_time:.4f} seconds to execute.\n")
        return result
    return wrapper
@time_logger
def YoutubeAudioDownload(video_url):
    video = YouTube(video_url.replace('www','music'))#
    audio = video.streams.filter(only_audio=True).order_by('abr').desc().first()#video.streams.filter(only_audio = True).first()
    author = video.author
    title = video.title
    thumbnail = video.thumbnail_url
    channel =video.channel_url
    id = video.channel_id
    set = {'author':author,'title':title,'channel':channel,'id':id, 'path':AUDIO_DOWNLOAD_DIR+title+".webm","thumbnail":thumbnail}
    try:
        audio.download(AUDIO_DOWNLOAD_DIR)
    except:
        print("Failed to download audio")

    print("audio was downloaded successfully")
    
    global dat
    dat.append(set)
    return dat 
@time_logger
def download_batch(urls):
    threads = []
    for url in urls:
        thread = threading.Thread(target=YoutubeAudioDownload, args=(url,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()
@time_logger
def batch_maker(urls):    
    batch_size = 4
    num_batches = len(urls) // batch_size
    batches = []
    for i in range(num_batches):
        batch = urls[i*batch_size : (i+1)*batch_size]
        batches.append(batch)
    return batches
```
